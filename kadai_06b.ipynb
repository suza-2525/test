## ç·´ç¿’3
å®Ÿéš›ã«å¤§æ–‡å­—ã¨å°æ–‡å­—ã‚’å¤‰æ›ã™ã‚‹æ™‚ã«ã¯ï¼Œã‚³ãƒ¼ãƒ‰ç•ªå·ã§ã¯ãªãï¼Œãƒ¡ã‚½ãƒƒãƒ‰ã‚’åˆ©ç”¨ã—ã¾ã™ï¼Python ã«ã¯æ–‡å­—åˆ—ã®å¤§æ–‡å­—ï¼Œå°æ–‡å­—ã¸ã®å¤‰æ›å‘½ä»¤ã¨ã—ã¦ï¼Œ```.upper()```, ```.lower()``` ãªã©ãŒã‚ã‚Šã¾ã™ï¼

ã‚ã‚‹ã‚¢ãƒ³ã‚±ãƒ¼ãƒˆã®å›ç­”ã§ï¼ŒYes, No ã‚’ç­”ãˆã¦ã‚‚ã‚‰ã„ã¾ã—ãŸãŒï¼Œ"YES", "Yes", "yes" ãªã©ï¼Œæ§˜ã€…ãªå›ç­”ãŒã‚ã‚Šã¾ã—ãŸï¼å¤§æ–‡å­—ã‚’å°æ–‡å­—ã«å¤‰æ›ã—ã¦ï¼ŒYes ã¨ç­”ãˆãŸäººã®å‰²åˆã‚’æ±‚ã‚ã¦ãã ã•ã„ï¼

````python
# ã‚¢ãƒ³ã‚±ãƒ¼ãƒˆçµæœã®ãƒªã‚¹ãƒˆ
answers = ["YES", "No", "yes", "Yes", "no", "yEs", "NO", "YeS"]

# ã™ã¹ã¦ã®å›ç­”ã‚’å°æ–‡å­—ã«å¤‰æ›
lowered = [a.lower() for a in answers]

# "yes" ã¨ç­”ãˆãŸäººã®æ•°ã‚’æ•°ãˆã‚‹
yes_count = lowered.count("yes")

# å‰²åˆï¼ˆï¼…ï¼‰ã‚’è¨ˆç®—
ratio = (yes_count / len(answers)) * 100

print(f"Yes ã¨ç­”ãˆãŸäººã®å‰²åˆã¯ {ratio:.1f}% ã§ã™ã€‚")
````

````python
bangou = 'f09f82a'
for i in range(1,15):
    p = str(hex(i))
    q = p.replace('0x','')
    print(q)
    b = bangou + q
    t = binascii.unhexlify(b).decode('utf-8')
    print(t)
````

Python ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã®å¿œç”¨ã§ã‚ˆãç”¨ã„ã‚‰ã‚Œã‚‹ã®ãŒï¼ŒPython ã‚’ä½¿ã£ã¦ã‚¤ãƒ³ã‚¿ãƒ¼ãƒãƒƒãƒˆä¸Šã®æƒ…å ±ã‚’è‡ªå‹•çš„ã«åé›†ã™ã‚‹ã“ã¨ã§ã™ï¼Web ãƒšãƒ¼ã‚¸ã‹ã‚‰å¿…è¦ãªæƒ…å ±ã ã‘ã‚’æŠ½å‡ºã—ã¦åˆ©ç”¨ã™ã‚‹ã“ã¨ã‚’ __ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°__ ã¨ã„ã„ã¾ã™ï¼ãã®ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°ã§ã‚ˆãä½¿ã‚ã‚Œã‚‹ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ä¸€ã¤ã« ```requests``` ãŒã‚ã‚Šã¾ã™ï¼

ä¾‹ãˆã°ï¼Œ[ã“ã®ãƒšãƒ¼ã‚¸](http://ogilab.kutc.kansai-u.ac.jp/complang/kadai7/page1.html) ã®å†…å®¹ã‚’ ```requests``` ã‚’ä½¿ã£ã¦å–å¾—ã™ã‚‹ã«ã¯ï¼Œä»¥ä¸‹ã®ã‚ˆã†ã«ã—ã¾ã™ï¼

````python
import requests

url = "http://ogilab.kutc.kansai-u.ac.jp/downloads/complang/kadai7/page1.html" #èª¿ã¹ãŸã„ãƒšãƒ¼ã‚¸ã®URL

r = requests.get(url)             # æŒ‡å®šã—ãŸ URL ã®ãƒšãƒ¼ã‚¸ã®æƒ…å ±ã‚’å–å¾—
r.encoding = r.apparent_encoding  # ãƒšãƒ¼ã‚¸ã®æ–‡å­—ã‚³ãƒ¼ãƒ‰ã‚’èª¿ã¹ã¦ï¼Œèª­ã¿è¾¼ã‚“ã ãƒšãƒ¼ã‚¸ã¸é©ç”¨ã™ã‚‹æ–‡å­—ã‚³ãƒ¼ãƒ‰ã¨ã—ã¦æŒ‡å®šã™ã‚‹
print("æ–‡å­—ã‚³ãƒ¼ãƒ‰ã¯", r.encoding)
print("----------------------")
print(r.text)                     # èª­ã¿è¾¼ã‚“ã ãƒšãƒ¼ã‚¸ã‚’è¡¨ç¤º
````

```r.apparent_encoding``` ã¨ã„ã†å‘½ä»¤ã¯ï¼Œè‡ªå‹•çš„ã«ãƒšãƒ¼ã‚¸ã«å«ã¾ã‚ŒãŸæ–‡å­—æƒ…å ±ã‹ã‚‰æ–‡å­—ã‚³ãƒ¼ãƒ‰ã‚’è§£æã—ï¼Œãã®çµæœã‚’è¿”ã—ã¦ãã‚Œã¾ã™ï¼ä»Šå›ã®å ´åˆã¯ ã€Œutf-8ã€ã¨ã„ã†çµæœãŒå‡ºã¦ã„ã¾ã™ã­ï¼ãã—ã¦ï¼Œèª­ã¿è¾¼ã‚“ã ãƒšãƒ¼ã‚¸ã®htmlã‚’ ```print(r.text)``` ã§è¡¨ç¤ºã•ã›ã¦è¦‹ã‚‹ã¨ï¼Œã¡ã‚ƒã‚“ã¨

## ç·´ç¿’5
ãã‚Œã§ã¯ï¼Œä¸Šè¨˜ã§æ–‡å­—åŒ–ã‘ã—ãŸãƒšãƒ¼ã‚¸(URL: http://ogilab.kutc.kansai-u.ac.jp/downloads/complang/kadai7/page3.html) ã«ã¤ã„ã¦ï¼ŒHTMLãƒ•ã‚¡ã‚¤ãƒ«ã®æ–‡å­—ã‚³ãƒ¼ãƒ‰ã¨ï¼ŒæŒ‡å®šã•ã‚ŒãŸæ–‡å­—ã‚³ãƒ¼ãƒ‰ charset ã‚’èª¿ã¹ã¦ã¿ã¦ãã ã•ã„ï¼

HTMLãƒ•ã‚¡ã‚¤ãƒ«ã¨ã—ã¦ã¯æ–‡å­—ã‚³ãƒ¼ãƒ‰ã¯ UTF-8 ã§ã‚ã‚‹ãŒï¼Œcharset ã§æŒ‡å®šã•ã‚ŒãŸæ–‡å­—ã‚³ãƒ¼ãƒ‰ã¯ SHIFT_JIS ã«ãªã£ã¦ã„ã‚‹ï¼

````python
import requests

url = " http://ogilab.kutc.kansai-u.ac.jp/downloads/complang/kadai7/page3.html" #èª¿ã¹ãŸã„ãƒšãƒ¼ã‚¸ã®URL

r = requests.get(url)             # æŒ‡å®šã—ãŸ URL ã®ãƒšãƒ¼ã‚¸ã®æƒ…å ±ã‚’å–å¾—
r.encoding = r.apparent_encoding  # ãƒšãƒ¼ã‚¸ã®æ–‡å­—ã‚³ãƒ¼ãƒ‰ã‚’èª¿ã¹ã¦ï¼Œèª­ã¿è¾¼ã‚“ã ãƒšãƒ¼ã‚¸ã¸é©ç”¨ã™ã‚‹æ–‡å­—ã‚³ãƒ¼ãƒ‰ã¨ã—ã¦æŒ‡å®šã™ã‚‹
print("æ–‡å­—ã‚³ãƒ¼ãƒ‰ã¯", r.encoding)
print("----------------------")
print(r.text)                     # èª­ã¿è¾¼ã‚“ã ãƒšãƒ¼ã‚¸ã‚’è¡¨ç¤º
````
