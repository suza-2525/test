# コンピュータの言語2025：課題12 クラスとオブジェクト指向

今回の授業ではクラスとオブジェクト指向について学びました．

__クラス__ とは，データとそれを扱う関数（メソッド）をひとまとまりにしたものです．

実際にクラスを作っていく練習をしましょう．

まずは初期化だけの機能を持つクラス ```Coordinate``` をつくってみます．

````python
class Coordinate:

    # クラスを使う時に最初に呼ばれる関数
    def __init__(self, x, y):
        self.x = x
        self.y = y
````

これを使ってみましょう．

````python
a = Coordinate(1,2)

print(f" a の x は {a.x} です")
print(f" a の y は {a.y} です")
````

次に，別の点との距離を計算する関数 ```distance``` を定義してみましょう

````python
class Coordinate:

    # クラスを使う時に最初に呼ばれる関数
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # 他の座標までの距離を計算する関数
    def distance(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        dd = dx**2 + dy**2
        return dd**0.5
````

そしてこれを使ってみましょう

````python
a = Coordinate(1,2)
b = Coordinate(4,6)

# a から b までの距離を計算
kyori = a.distance(b)

print( f"a から b までの距離は {kyori} です")
````

さて，このクラスに対して ```print``` をしてみるとどうなるでしょうか？

````python
print(a)
````

デフォルトの表示は、オブジェクトの型とメモリ上のアドレスを示すだけで、あまり有益ではありません。そこで，```print``` の命令に対して，わかりやすくデータが表示されるように，```__str__``` 関数を追加します．

````python
class Coordinate:

    # クラスを使う時に最初に呼ばれる関数
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # 他の座標までの距離を計算する関数
    def distance(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        dd = dx**2 + dy**2
        return dd**0.5

    # print()命令で呼ばれる関数
    def __str__(self):
        return f"座標は ({self.x}, {self.y}) です"
````

````python
a = Coordinate(1,2)
print(a)
````

## 課題1
以下のように使える犬のクラス ```Dog``` を作成してください．

```
p = Dog("ポチ", 2)
p.bark()   # 「わんわん，ぼくはポチだわん」と出力

s = Dog("シロ", 3)
s.back()  #　「わんわん，ぼくはシロだわん」と出力

p.play(s) # 「シロは大事なお兄ちゃん」と出力
s.play(p) # 「ポチは大事な弟ちゃん」と出力
```

````python
class Dog:

    def __init__(self, name, year):
        self.name = name
        self.year = year

    def bark(self):
        p = self.name
        print(f"わんわん，ぼくは{p}だわん")

    def play(self, other):
        if self.year<other.year:
            print("シロは大事なお兄ちゃん")
        else:
            print("ポチは大事な弟ちゃん")
````

````python
p = Dog("ポチ", 2)
p.bark()   # 「わんわん，ぼくはポチだわん」と出力
````

````python
s = Dog("シロ", 3)
s.bark()   # 「わんわん，ぼくはポチだわん」と出力
````

````python
p.play(s) # 「シロは大事なお兄ちゃん」と出力
s.play(p) # 「ポチは大事な弟ちゃん」と出力
````

## 課題2

以下のように使える長方形のクラス ```Rect``` を作成して下さい．

```
rect = Rect()
rect.set_tate(10)    # 縦の長さを設定
rect.set_yoko(20)    # 横の長さを設定
print(rect)          # "縦:10, 横:20" と表示
S = rect.calc_menseki()
print(f"面積は{S}です")      
L = rect.calc_nagasa()  
print(f"周囲の長さは{L}です")
```


````python
class Rect:
    def __init__(self):
        self.x = 0
        self.y = 0

    def set_tate(self, x):
        self.x = x

    def set_yoko(self, y):
        self.y = y

    def calc_menseki(self):
        S = self.x * self.y
        return S
    def calc_nagasa(self):
        L = self.x * 2 + self.y * 2
        return L
    def __str__(self):
        return f"縦:{self.x}, 横:{self.y}"

######################################
rect = Rect()
rect.set_tate(10)    # 縦の長さを設定
rect.set_yoko(20)    # 横の長さを設定
print(rect)          # "縦:10, 横:20" と表示
S = rect.calc_menseki()
print(f"面積は{S}です")
L = rect.calc_nagasa()
print(f"周囲の長さは{L}です")
````
